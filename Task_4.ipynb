This is for the training dataset. Negative tweets are the highest in volume compared to the other types.

```python
import pandas as pd
from textblob import TextBlob

# Load the social media data
df = pd.read_csv('twitter_training.csv')

# Perform sentiment analysis on the text data
'''df['SENTIMENT'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)'''

# Visualize the sentiment distribution
import matplotlib.pyplot as plt
plt.hist(df['SENTIMENT'], bins=20)
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.title('Sentiment Distribution of Social Media Data')
plt.show()


# Identify topics or brands with the most positive or negative sentiment
top_positive = df.sort_values('SENTIMENT', ascending=False).head(10)
top_negative = df.sort_values('SENTIMENT').head(10)

print('Top 10 Most Positive Topics/Brands:')
print(top_positive)

print('\nTop 10 Most Negative Topics/Brands:')
print(top_negative)
```
